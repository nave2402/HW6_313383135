```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import seaborn as sns
```

```python
df = pd.read_csv('LP.csv')
# 1) counts the number of words in the description of the city.
df['num_of_words'] = df['Description'].str.count(' ') + 1

# 2) remove_punctuations

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
df['No_punct_description'] = df['Description'].apply(remove_punctuations)

# 3) restaurants_counter
df['has_restaurants'] = df['No_punct_description'].str.count('restaurant')
  
# 4) museums_counter
df['has_museums'] = df['No_punct_description'].str.count('museum')

# 5) has_beaches
df['has_beaches'] = df['No_punct_description'].str.count('beach') + df['No_punct_description'].str.count('sea') + df['No_punct_description'].str.count('ocean')
```

```python
# 6) average number of words in the description column 
df['num_of_words'].mean()
```

```python
# 7) the city with the longest description column 
df['City'][df['num_of_words'] == df['num_of_words'].max()]
```

```python
# 8) Draw a histogram of the column “num_of_words” 
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
sns.histplot(data=df,x='num_of_words',bins=30,kde=True)
plt.xlabel("Number of Words")
plt.title("Number of Words - Histogram")
```

```python
# 9) Draw a scatter plot that shows the relationship between number of beaches and number of restaurants. 
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
sns.scatterplot(x='has_beaches',y='has_restaurants',data=df)
plt.xlabel("Number of Beaches")
plt.ylabel("Number of Restaurants")
plt.title("The Relationship Between Number of Beaches and Number of Restaurants - scatter plot")
```

```python
# 11) Draw a scatter plot that shows the relationship between number of museums and the number of beaches. 
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
sns.scatterplot(x='has_beaches',y='has_museums',data=df)
plt.xlabel("Number of Beaches")
plt.ylabel("Number of Museums")
plt.title("The Relationship Between Number of Beaches and Number of Museums - scatter plot")
```
